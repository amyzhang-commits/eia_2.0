````python
#!/usr/bin/env python3
"""
dataset_merger.py

Two-phase ingestion module for Excel/CSV datasets.

Phase 1: Schema Preview + Memory Estimation (NO merging)
         Column schemas, row counts, memory footprint, merge recommendations.
Phase 2: Merge files with coverage analysis and optional heatmap.

Designed for reproducibility, portability, and large public datasets (e.g. EIA).

Functions
---------
preview_folder_schema(folder_path=None)
    Inspect schemas + estimate memory usage without loading data.
merge_folder_files(folder_path=None, save=False, show_heatmap=True)
    Memory-safe merge with coverage heatmap visualization.
"""

from pathlib import Path
from typing import Optional, Union
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.colors import ListedColormap
from tqdm import tqdm
import openpyxl
import psutil

# ======================================================
# 📁 Project-relative paths
# ======================================================

FALLBACK_DATASETS_PATH = Path(__file__).parent.parent / "datasets_thermoelectric"
DATASETS_OUTPUT = Path(__file__).parent.parent / "datasets_merged"
DATASETS_OUTPUT.mkdir(exist_ok=True)

# ======================================================
# 🧠 Header detection logic
# ======================================================

def detect_header_row(
    file_path: Union[str, Path],
    nrows: int = 10,
    string_ratio_threshold: float = 0.5
) -> int:
    """Detect the most likely header row in a CSV or Excel file."""
    file_path = Path(file_path)
    
    if file_path.suffix.lower() == ".xlsx":
        preview = pd.read_excel(file_path, header=None, nrows=nrows)
    elif file_path.suffix.lower() == ".csv":
        preview = pd.read_csv(file_path, header=None, nrows=nrows)
    else:
        raise ValueError(f"Unsupported file type: {file_path.name}")

    for i, row in preview.iterrows():
        string_ratio = row.apply(lambda x: isinstance(x, str)).mean()
        if string_ratio > string_ratio_threshold:
            return i
    return 0

# ======================================================
# 📄 File row counters
# ======================================================

def get_total_rows_excel(file_path: Union[str, Path], header_row: int = 0) -> int:
    """Get exact data row count from Excel file (excluding header)."""
    file_path = Path(file_path)
    wb = openpyxl.load_workbook(file_path, read_only=True)
    sheet = wb.active
    total_rows = sheet.max_row - header_row
    wb.close()
    return total_rows

def get_total_rows_csv(file_path: Union[str, Path], header_row: int = 0) -> int:
    """Get exact data row count from CSV file (excluding header)."""
    file_path = Path(file_path)
    with open(file_path, "r", encoding="utf-8", errors="ignore") as fin:
        total_lines = sum(1 for _ in fin)
    return total_lines - 1  # Subtract header row

# ======================================================
# 📄 Unified file reader
# ======================================================

def read_file(file_path: Union[str, Path]) -> pd.DataFrame:
    """Read a single file with auto-detected header."""
    file_path = Path(file_path)
    header_row = detect_header_row(file_path)
    print(f"Reading {file_path.name} → detected header row: {header_row}")

    if file_path.suffix.lower() == ".xlsx":
        return pd.read_excel(file_path, header=header_row)
    elif file_path.suffix.lower() == ".csv":
        return pd.read_csv(file_path, header=header_row)

# ======================================================
# 🔍 Phase 1: Schema Preview + Memory Estimation
# ======================================================

def preview_folder_schema(folder_path: Optional[Union[str, Path]] = None) -> None:
    """
    Inspect all files without merging. Provides schema analysis and memory estimation.
    
    Outputs:
    - Column counts and header detection per file
    - Schema consistency warnings
    - Memory footprint estimation
    - Merge safety recommendations

    Parameters
    ----------
    folder_path : str or pathlib.Path, optional
        Folder containing CSV/Excel files. If None, uses FALLBACK_DATASETS_PATH.
    """
    folder = Path(folder_path) if folder_path else FALLBACK_DATASETS_PATH
    files = [f for f in folder.glob("*") 
             if f.suffix.lower() in [".xlsx", ".csv"] and not f.name.startswith(".")]

    if not files:
        raise ValueError(f"No valid files found in {folder}")

    print(f"\n🔍 Previewing schema in folder: {folder}\n")
    schemas = {}
    row_counts = {}
    mem_estimates_gb = {}

    for f in files:
        header_row = detect_header_row(f)
        
        # Get columns (zero rows = headers only)
        if f.suffix.lower() == ".xlsx":
            cols = pd.read_excel(f, header=header_row, nrows=0).columns.tolist()
            total_rows = get_total_rows_excel(f, header_row)
        else:
            cols = pd.read_csv(f, header=header_row, nrows=0).columns.tolist()
            total_rows = get_total_rows_csv(f, header_row)

        schemas[f.name] = cols
        row_counts[f.name] = total_rows
        
        # Memory estimate: sample 100 rows → extrapolate
        if f.suffix.lower() == ".xlsx":
            sample = pd.read_excel(f, header=header_row, nrows=100)
        else:
            sample = pd.read_csv(f, header=header_row, nrows=100)
        
        sample_bytes = sample.memory_usage(deep=True).sum()
        est_bytes = sample_bytes * (total_rows / 100)
        est_gb = est_bytes / 1e9
        mem_estimates_gb[f.name] = est_gb
        
        print(f"  {f.name}: {len(cols)} cols | header: {header_row} | "
              f"{total_rows:,} rows | ~{est_gb:.2f} GB")

    # Schema consistency
    all_columns = set().union(*schemas.values())
    print(f"\n📊 Total unique columns across files: {len(all_columns)}")
    
    for name, cols in schemas.items():
        missing = all_columns - set(cols)
        if missing:
            print(f"  ⚠️  {name} missing: {sorted(list(missing))}")

    # Memory recommendation
    total_est_gb = sum(mem_estimates_gb.values())
    system_ram_gb = psutil.virtual_memory().total / 1e9
    
    print(f"\n💾 Estimated total memory: {total_est_gb:.2f} GB")
    print(f"🖥️  Available system RAM:   {system_ram_gb:.2f} GB")
    
    if total_est_gb > 0.7 * system_ram_gb:
        print("⚠️   RECOMMENDATION: Use chunked/incremental merge (total exceeds 70% RAM)")
    else:
        print("✅   RECOMMENDATION: Direct merge safe")
    
    print("\n✅ Phase 1 complete. Review before merging.\n")

# ======================================================
# 🔗 Phase 2: Merge + Coverage Summary
# ======================================================

def merge_folder_files(
    folder_path: Optional[Union[str, Path]] = None,
    save: bool = False,
    show_heatmap: bool = True
) -> pd.DataFrame:
    folder = Path(folder_path) if folder_path else FALLBACK_DATASETS_PATH
    files = [
        f for f in folder.glob("*")
        if f.suffix.lower() in [".xlsx", ".csv"] and not f.name.startswith(".")
    ]

    if not files:
        raise ValueError(f"No valid files found in {folder}")

    print(f"\n🔗 Merging files from: {folder}")

    merged_df = pd.concat(
        (read_file(f) for f in tqdm(files, desc="Merging files")),
        ignore_index=True
    )

    print(f"\n📐 Merged DataFrame shape: {merged_df.shape}")
    print(f"✅ {len(files)} files merged\n")

    # Coverage summary
    absent_count = merged_df.isnull().sum()
    absent_pct = (absent_count / len(merged_df)) * 100
    coverage_df = pd.DataFrame({
        "Absent Count": absent_count,
        "Absent Percentage": absent_pct
    }).loc[merged_df.columns]

    with pd.option_context("display.max_rows", None):
        print("📊 Coverage Summary:\n")
        print(coverage_df.to_string())
        print()

    # Heatmap
    if show_heatmap:
        absent_mask = merged_df.isnull()
        cmap = ListedColormap(["#00008B", "#D3D3D3"])  # present, absent

        plt.figure(figsize=(16, 8))
        sns.heatmap(absent_mask, cmap=cmap, cbar=False, yticklabels=False)
        plt.title("Coverage Heatmap")
        plt.xlabel("Columns")
        plt.ylabel("Rows")

        timestamp = pd.Timestamp.now().strftime("%Y%m%d_%H%M%S")
        save_path = DATASETS_OUTPUT / f"coverage_heatmap_{folder.name}_{timestamp}.png"
        plt.savefig(save_path, dpi=300, bbox_inches="tight")
        print(f"✅ Coverage heatmap saved to: {save_path}\n")
        plt.show()

    if save:
        timestamp = pd.Timestamp.now().strftime("%Y%m%d_%H%M%S")
        output_file = DATASETS_OUTPUT / f"merged_{folder.name}_{timestamp}.csv"
        merged_df.to_csv(output_file, index=False)
        print(f"💾 Merged CSV saved to: {output_file}\n")

    # Top 10 columns with most absence
    top_missing = (
        coverage_df
        .sort_values("Absent Count", ascending=False)
        .head(10)
    )

    print("🔟 Top 10 Columns with Most Missing Values:\n")
    print(top_missing.to_string())
    print()

    return merged_df


# ======================================================
# 🏁 Script execution
# ======================================================

if __name__ == "__main__":
    preview_folder_schema()
    merge_folder_files()
````

# Step 2: MEMORY ISSUE BEFORE MERGE? Commit to chunking/incremental merging (in the future???)

Yesss, this is a classic “big data in Python without melting your laptop” trick 🫣🔥

Right now, your `merge_folder_files()` does this:

```python
merged_df = pd.concat(
    (read_file(f) for f in tqdm(files, desc="Merging files")),
    ignore_index=True
)
```

…and `read_file(f)` **loads the entire file into memory**, then `pd.concat` creates **a new big DataFrame** in memory, which temporarily doubles memory usage. For 917k × 70 it’s fine, but if your files get bigger, this can spike RAM.

We can improve memory usage a couple of ways:

---

### 1️⃣ **Incremental concatenation (reuse memory)**

Instead of building a huge generator and then concatenating all at once:

```python
merged_df = pd.DataFrame()
for f in tqdm(files, desc="Merging files"):
    df = read_file(f)
    merged_df = pd.concat([merged_df, df], ignore_index=True)
    del df  # free temporary DataFrame memory
```

✅ Pros:

* You free the temporary `df` after each file.
* Good if your files are moderately sized.

❌ Cons:

* `pd.concat` in a loop **can still be slow** because each concat copies memory.

---

### 2️⃣ **List-of-DataFrames + single concat (current approach)**

```python
merged_df = pd.concat([read_file(f) for f in files], ignore_index=True)
```

* Already memory efficient for **moderate numbers of files**.
* If files are huge, memory spike occurs during `pd.concat`.

---

### 3️⃣ **Chunked reading (best for HUGE files)**

If a single Excel/CSV is massive:

```python
chunks = []
for f in tqdm(files, desc="Merging files in chunks"):
    if f.suffix.lower() == ".csv":
        for chunk in pd.read_csv(f, chunksize=50000):  # 50k rows at a time
            chunks.append(chunk)
    elif f.suffix.lower() == ".xlsx":
        chunks.append(pd.read_excel(f))
merged_df = pd.concat(chunks, ignore_index=True)
```

✅ Pros:

* Never loads a giant file all at once (for CSVs).
* Keeps memory spikes manageable.

❌ Cons:

* Excel chunking isn’t native, so huge `.xlsx` files still load fully.

---

### 4️⃣ **Optional: immediately write merged chunks to disk**

Instead of holding everything in memory, you can:

```python
output_file = DATASETS_OUTPUT / "merged_temp.csv"
for f in files:
    df = read_file(f)
    df.to_csv(output_file, mode="a", index=False, header=not output_file.exists())
    del df
```

* This way, **memory never holds all files together**.
* Only the **final CSV** is written.
* Super scalable for massive datasets.

---

💡 My recommendation for your EIA workflow:

* For now, your current approach is fine — `917k x 70` is fine for memory.
* If you want to be future-proof: implement **incremental concat** or **chunked CSV writing**.
* Avoid reading multiple Excel files fully if they grow >1M rows each.

Great question! Let me break down how encoding error handling works:

## The three strategies

When Python encounters a byte sequence it can't decode as UTF-8:

### 1. **`errors="strict"`** (default)
```python
pd.read_csv("file.csv", encoding="utf-8")  # errors="strict" is default
```
- **Behavior**: Raises `UnicodeDecodeError` and stops
- **Use case**: When data integrity is critical and you want to know immediately if there's a problem
- **Example**: Byte `\x93` (left smart quote in Windows-1252) → **CRASH**

### 2. **`errors="ignore"`**
```python
pd.read_csv("file.csv", encoding="utf-8", errors="ignore")
```
- **Behavior**: Silently skips/removes problematic bytes
- **Use case**: When you just want it to work and don't care about lost characters
- **Example**: `"Café\x93"` → becomes `"Café"` (the `\x93` disappears)
- **Risk**: Data loss is invisible—you won't know something was corrupted

### 3. **`errors="replace"`**
```python
pd.read_csv("file.csv", encoding="utf-8", errors="replace")
```
- **Behavior**: Replaces bad bytes with the replacement character `�` (U+FFFD)
- **Use case**: When you want processing to continue but need visibility into corruption
- **Example**: `"Café\x93"` → becomes `"Café�"`
- **Benefit**: You can see in your data that something went wrong

## Visual comparison

Imagine a CSV with plant name: `José's Plant` encoded in Windows-1252:

```
Strict:   CRASH ❌
Ignore:   "Joss Plant" (apostrophe vanishes silently)
Replace:  "Jos�s Plant" (� shows something was wrong)
```

## My recommendation for your script

Use **`errors="replace"`** in `read_file()`:
```python
elif file_path.suffix.lower() == ".csv":
    return pd.read_csv(file_path, header=header_row, 
                      encoding="utf-8", errors="replace")
```

**Why?** 
- Processing continues (unlike `strict`)
- Problems are visible (unlike `ignore`)  
- You can later search for `�` characters to identify affected records
- For EIA-type datasets, a few corrupted special characters won't invalidate the entire merge

The script already uses `errors="ignore"` in `get_total_rows_csv()`, which is fine there since you're just counting lines. But for actual data loading, `replace` is safer.
